```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Academic_Performance.csv")
df
```

```python
df.isnull().sum()
```

```python
avg_math = int(df['math_score'].mean())
df['math_score'].fillna(avg_math, inplace=True)
df
```

```python
avg_reading = int(df['reading_score'].mean())
df['reading_score'].fillna(avg_reading, inplace=True)
df
```

```python
avg_writing = int(df['writing_score'].mean())
df['writing_score'].fillna(avg_writing, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
col = ['math_score','reading_score','writing_score','placement_score']
df.boxplot(col)
```

```python
upper_bound = 80
lower_bound = 60
col = df['math_score']
math_outliers = []
for i in col:
    if i > upper_bound or i < lower_bound:
        math_outliers.append(int(i))
math_outliers
```

```python
df['math_score'].replace(math_outliers, avg_math, inplace=True)
df
```

```python
from scipy import stats
z = np.abs(stats.zscore(df['reading_score']))
col = df['math_score']
threshold = 0.65
sample_outlier = np.where(z>threshold)
sample_outlier
```

```python
out1 = int(df["reading_score"].loc[3])
out2 = int(df["reading_score"].loc[22])
out3 = int(df["reading_score"].loc[27])
reading_outliers = [out1, out2, out3]
reading_outliers
```

```python
df['reading_score'].replace(reading_outliers, avg_reading, inplace=True)
df
```

```python
writing_outliers = []
sorted_rscore = sorted(df['writing_score'])

q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)

IQR = q3 - q1

lower_bound = q1 - (1.5*IQR)
upper_bound = q3 + (1.5*IQR)

for i in sorted_rscore:
    if (i< lower_bound or i > upper_bound):
        writing_outliers.append(int(i))

print(writing_outliers)
```

```python
df['writing_score'].replace(writing_outliers, avg_writing, inplace=True)
df
```

```python
fig, ax = plt.subplots(figsize = (18,10))
ax.scatter(df['placement_score'], df['placement_offer_count'])
plt.show()
```

```python
placement_outliers = [20,60]
placement_outliers
```

```python
avg_placement = int(df['placement_score'].mean())
df['placement_score'].replace(placement_outliers, avg_placement, inplace=True)
df
```
