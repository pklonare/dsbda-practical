```python
avg_stroke = round(df["stroke"].mean(),2)
print("Average of stroke : ",avg_stroke)
df['stroke'].fillna(avg_stroke, inplace=True)
```

```python
avg_hp = int(df["horsepower"].mean())
print("Average of horsepower : ",avg_hp)
df['horsepower'].fillna(avg_hp, inplace=True)
```

```python
avg_prpm = int(df["peak-rpm"].mean())
print("Average of peak rpm : ",avg_prpm)
df['peak-rpm'].fillna(avg_prpm, inplace=True)
```

```python
df.dropna(subset=["horsepower-binned"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.isnull().sum()
```
